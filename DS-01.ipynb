```python
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv("tips.csv")


plt.bar(data['size'], data['tip'])

plt.title("Bar Chart")

# Setting the X and Y labels
plt.xlabel('size')
plt.ylabel('Tip')

# Adding the legends
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv("tips.csv")

# histogram of total_bills
plt.hist(data['tip'])

plt.title("Histogram")

# Adding the legends
plt.show()
```
